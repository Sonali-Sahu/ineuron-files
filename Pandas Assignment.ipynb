```python
#Q9. How do you concatenate two Pandas DataFrames?

# create df1
student_dict = {"name":["ABC","DEF","GHI"],"roll_num" :[100,200,300],"Class":[4,6,8]}
student_df = pd.DataFrame(student_dict)
#print(student_df)

# create df2 for axis =1 concat

school_dict = {"school":["DAV","MAV","Vincent"]}
school_df = pd.DataFrame(school_dict)

# horizontal concatenation
horizontal_concate = pd.concat([student_df,school_df],axis = 1)
print(horizontal_concate)

print()

# new student details
student_dict1 = {"name":["JKL"],"roll_num" :[400],"Class":[10]}
student_df1 = pd.DataFrame(student_dict1)

# horizontal concatenation
vertical_concate = pd.concat([student_df,student_df1],axis = 0)
print(vertical_concate)
```

```python
#Q23. How do you create a new column in a DataFrame based on the values of another column?

def return_grade_category(df_row):
    if df_row["grade"] == 'O' or df_row["grade"] == "E" :
        return "Excellent"
    elif df_row["grade"] == "A" or df_row["grade"] == "B":
        return "good"
    elif df_row["grade"] == "C":
        return "Average"
    else:
        return "Failed"
    
example_df['grade_category'] = example_df.apply(lambda df_row: return_grade_category(df_row), axis=1)

example_df
```

# Q25. What is the difference between .loc and .iloc in Pandas?
Ans. .loc function is used to locate rows/columns based on some conditions or select rows for a give range

 .iloc is used to get given indices rows or some chunk of data for a given range for rows and columns
